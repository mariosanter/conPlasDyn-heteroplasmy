```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Plasmid Fusion model: Simulation Widget 

```python
def interactive_dimerization_model():

    n_slider = widgets.IntSlider(
        value=4,
        min=2,
        max=12,
        step=2,
        description="n",
        continuous_update=False
    )

    ddim_slider = widgets.FloatSlider(
        value=0.01,
        min=0.0,
        max=1.0,
        step=0.01,
        description="ddim",
        continuous_update=False
    )

    dmon_slider = widgets.FloatSlider(
        value=0.0,
        min=0.0,
        max=1.0,
        step=0.01,
        description="dmon",
        continuous_update=False
    )

    gen_slider = widgets.IntSlider(
        value=80,
        min=1,
        max=200,
        step=1,
        description="gen",
        continuous_update=False
    )

    f_slider = widgets.FloatSlider(
        value=0.1,
        min=0.0,
        max=1.0,
        step=0.01,
        description="f",
        continuous_update=False
    )

    output = widgets.Output()

    def update(change=None):
        with output:
            clear_output(wait=True)

            dM = dimerizationModel(
                n=n_slider.value,
                ddim=ddim_slider.value,
                dmon=dmon_slider.value
            )

            dM.growth(gen_slider.value, f_slider.value)
            dM.growth_groupPlot()

            plt.show()

    for w in [n_slider, ddim_slider, dmon_slider, gen_slider, f_slider]:
        w.observe(update, names="value")

    display(
        widgets.VBox([
            widgets.HBox([n_slider, gen_slider]),
            widgets.HBox([ddim_slider, dmon_slider]),
            f_slider,
            output
        ])
    )

    update()

interactive_dimerization_model()
```

#### Plasmid Fusion model: Simulation & Plots 

```python
dM=dimerizationModel(4,0.1,.1)
dM.growth(85,0.1)
ggP=dM.growth_groupPlot()
dM=dimerizationModel(6,0.1,.1)
dM.growth(85,0.1)
dM.growth_groupPlot(plot=ggP.plot,alpha=1.,vivid=.3,brighter=1.5,hue=.2,zorder=1)
ggP.plot.fig.savefig("manuscript/dimerization_low_eq.pdf")
ggP.plot.fig
```
